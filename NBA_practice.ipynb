# Getting To Know The Commands

```python
# Install Dependencies
from nba_api.stats.endpoints import playercareerstats # Provides single player career stats"
from nba_api.live.nba.endpoints import scoreboard # Provides scoreboards for that day"
from nba_api.stats.static import players # Receive player information i.e. player_id
from nba_api.stats.static import teams
from nba_api.stats.endpoints import leagueleaders
import pandas as pd
import numpy as np
```

```python
# Practice with pulling nikola jokic career stats
career = playercareerstats.PlayerCareerStats(player_id='203999')
```

```python
# Pandas Data Frames
career.get_data_frames()[0]
```

```python
# Pull all of the active players player_id
player_dict = players.get_players()

# Turn dictionary to a dataframe
players = pd.DataFrame.from_dict(player_dict)

# Show the player_dict
players
```

```python
# Get team dict
teams = teams.get_teams()

# Turn dictionary to dataframe
teams = pd.DataFrame.from_dict(teams)

# Show the teams
teams.head()
```

```python
# Get active players only
active_players = players.loc[players['is_active'] == True]

# Show data
active_players.head()
```

```python
# Get all players for the 22-23 season using league leaders
top_600 = leagueleaders.LeagueLeaders(
    season='2003-04',
    season_type_all_star='Regular Season',
    stat_category_abbreviation='PTS'
).get_data_frames()[0][:600]
# Show the top 50 players
top_600.head()
```

```python
# View all the columns
top_600.dtypes
```

```python
# Remove unnecessary columns for graphing
NBA_PLAYERS = top_600[['PLAYER', 'RANK', 'TEAM', 'GP', 'MIN', 'FGM', 'FGA', 'FG_PCT', 'FG3M', 'FG3A',
                      'FG3_PCT', 'FTM', 'FTA', 'FT_PCT', 'OREB', 'DREB', 'REB', 'AST', 'STL', 'BLK',
                      'TOV', 'PF', 'PTS', 'EFF', 'AST_TOV', 'STL_TOV']]

# Show the data
NBA_PLAYERS.head()
```

```python
# Download the data
NBA_PLAYERS.to_csv('NBA_PLAYERS.csv')
```
